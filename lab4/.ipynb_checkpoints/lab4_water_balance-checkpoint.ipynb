```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv('data/water_potability.csv')
df.describe()
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df['Potability'].value_counts().plot(kind = 'bar')
```

```python
# Данные не сбалансированы по целевой переменной
```

```python
min_ratio = df['Potability'].value_counts().min()
min_ratio
```

```python
df_0 = df[df['Potability'] == 0].sample(min_ratio, replace = True)
df_1 = df[df['Potability'] == 1]
df_balanced = pd.concat([df_0, df_1])
```

```python
df_balanced['Potability'].value_counts().plot(kind = 'bar')
```

```python
# теперь норм
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df_balanced.corr(), annot = True)
```

```python
sns.scatterplot(x = 'Organic_carbon', y = 'ph', data = df_balanced, hue = 'Potability'  )
```

```python
# нормализуем данные
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
df_scaled_vales = scaler.fit_transform(df_balanced)
df_scaled = pd.DataFrame(columns = list(df.columns), data = df_scaled_vales)
```

```python
df_scaled.head()
```

```python
for col_name in df.columns:
    unique,counts=np.unique(df_scaled[col_name],return_counts=True)
    plt.figure(figsize=(12,6))
    plt.title(col_name)
    plt.bar(unique,counts)
    plt.ylabel('Число случаев')
    plt.xlabel('Возможные случаи')
    plt.show()
```

```python
y=df_scaled.pop('Potability')
```

```python
x=df_scaled.values
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Создадим наш перцептрон
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
```

```python
model = Perceptron(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
acc = accuracy_score(y_pred, y_test)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

print('Accuracy:', acc)
print('F-мера:', f1)
print('ROC AUC:', roc_auc)
```
